# Titanic Exploratory Data Analysis

This notebook performs a quick EDA on the provided `data/data_train.csv` sample.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
train = pd.read_csv('../data/data_train.csv')
train.head()
```

```python
# Basic info
train.info()
```

```python
# Summary statistics
train.describe(include='all')
```

```python
# Visual: Survival count
sns.countplot(x='Survived', data=train)
plt.title('Survival Count (sample)')
plt.show()
```

```python
# Visual: Age distribution
train['Age'].plot(kind='hist', bins=10)
plt.title('Age distribution (sample)')
plt.xlabel('Age')
plt.show()
```

```python
# Correlation heatmap (numeric columns)
num = train.select_dtypes(include=['float64','int64'])
sns.heatmap(num.corr(), annot=True)
plt.title('Correlation (sample)')
plt.show()
```
